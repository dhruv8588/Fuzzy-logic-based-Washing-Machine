```python
import numpy as np
import skfuzzy as fuzz
import skfuzzy.membership as mf
import matplotlib.pyplot as plt
```

## Input and Output

```python
x_dirtinessOfClothes = np.arange(0, 101)
x_typeOfDirt = np.arange(0, 101)
x_volOfClothes = np.arange(0, 101)
x_typeOfFabric = np.arange(0, 101)

y_washingTime = np.arange(0, 101)
y_washingSpeed = np.arange(0, 1201)
y_waterIntake = np.arange(0, 101)
y_waterTemp = np.arange(0, 81)
```

## Membership function of input and output

```python
dirtinessOfClothes_small = mf.trimf(x_dirtinessOfClothes, [0, 0, 70])
dirtinessOfClothes_med = mf.trimf(x_dirtinessOfClothes, [0, 50, 100])
dirtinessOfClothes_large = mf.trimf(x_dirtinessOfClothes, [10, 100, 101])

typeOfDirt_notgreasy = mf.trimf(x_typeOfDirt, [0, 0, 80])
typeOfDirt_med = mf.trimf(x_typeOfDirt, [0, 40, 100])
typeOfDirt_greasy = mf.trimf(x_typeOfDirt, [20, 100, 101])

volOfClothes_small = mf.trimf(x_volOfClothes, [0, 0, 60])
volOfClothes_med = mf.trimf(x_volOfClothes, [0, 60, 100])
volOfClothes_large = mf.trimf(x_volOfClothes, [20, 100, 101])

typeOfFabric_silk = mf.trimf(x_typeOfFabric, [0, 10, 50])
typeOfFabric_woolen = mf.trimf(x_typeOfFabric, [0, 50, 100])
typeOfFabric_cotton = mf.trimf(x_typeOfFabric, [10, 80, 100])

washingTime_veryshort = mf.trimf(y_washingTime, [0, 0, 40])
washingTime_short = mf.trimf(y_washingTime, [0, 20, 60])
washingTime_med = mf.trimf(y_washingTime, [0, 50, 100])
washingTime_long = mf.trimf(y_washingTime, [20, 70, 100])
washingTime_verylong = mf.trimf(y_washingTime, [30, 100, 101])

washingSpeed_veryslow = mf.trimf(y_washingSpeed, [0, 0, 500])
washingSpeed_slow = mf.trimf(y_washingSpeed, [0, 300, 700])
washingSpeed_med = mf.trimf(y_washingSpeed, [0, 600, 1200])
washingSpeed_fast = mf.trimf(y_washingSpeed, [100, 900, 1200])
washingSpeed_veryfast = mf.trimf(y_washingSpeed, [400, 1200, 1201])

waterIntake_little = mf.trimf(y_waterIntake, [0, 0, 50])
waterIntake_normal = mf.trimf(y_waterIntake, [0, 50, 100])
waterIntake_alotof = mf.trimf(y_waterIntake, [30, 100, 101])

waterTemp_low = mf.trimf(y_waterTemp, [0, 0, 60])
waterTemp_normal = mf.trimf(y_waterTemp, [0, 40, 80])
waterTemp_high = mf.trimf(y_waterTemp, [10, 80, 81])
```

## Plot of membership function (input & output)

```python
fig, (ax0, ax1, ax2, ax3, ax4, ax5, ax6, ax7) = plt.subplots(nrows=8, figsize=(5, 20))

ax0.plot(x_dirtinessOfClothes, dirtinessOfClothes_small, 'r', linewidth=2, label='Small')
ax0.plot(x_dirtinessOfClothes, dirtinessOfClothes_med, 'g', linewidth=2, label='Medium')
ax0.plot(x_dirtinessOfClothes, dirtinessOfClothes_large, 'b', linewidth=2, label='Large')
ax0.set_title('Dirtiness of Clothes')
ax0.legend()

ax1.plot(x_typeOfDirt, typeOfDirt_notgreasy, 'r', linewidth=2, label='Not greasy')
ax1.plot(x_typeOfDirt, typeOfDirt_med, 'g', linewidth=2, label='Medium')
ax1.plot(x_typeOfDirt, typeOfDirt_greasy, 'b', linewidth=2, label='Greasy')
ax1.set_title('Type of Dirt')
ax1.legend()

ax2.plot(x_volOfClothes, volOfClothes_small, 'r', linewidth=2, label='Small')
ax2.plot(x_volOfClothes, volOfClothes_med, 'g', linewidth=2, label='Medium')
ax2.plot(x_volOfClothes, volOfClothes_large, 'b', linewidth=2, label='Large')
ax2.set_title('Volume of Clothes')
ax2.legend()

ax3.plot(x_typeOfFabric, typeOfFabric_silk, 'r', linewidth=2, label='Silk')
ax3.plot(x_typeOfFabric, typeOfFabric_woolen, 'g', linewidth=2, label='Woolen')
ax3.plot(x_typeOfFabric, typeOfFabric_cotton, 'b', linewidth=2, label='Cotton')
ax3.set_title('Type of Fabric')
ax3.legend()

ax4.plot(y_washingTime, washingTime_veryshort, 'r', linewidth=2, label='Very short')
ax4.plot(y_washingTime, washingTime_short, 'g', linewidth=2, label='Short')
ax4.plot(y_washingTime, washingTime_med, 'b', linewidth=2, label='Medium')
ax4.plot(y_washingTime, washingTime_long, 'y', linewidth=2, label='Long')
ax4.plot(y_washingTime, washingTime_verylong, 'm', linewidth=2, label='Very long')
ax4.set_title('Washing Time')
ax4.legend()

ax5.plot(y_washingSpeed, washingSpeed_veryslow, 'r', linewidth=2, label='Very slow')
ax5.plot(y_washingSpeed, washingSpeed_slow, 'g', linewidth=2, label='Slow')
ax5.plot(y_washingSpeed, washingSpeed_med, 'b', linewidth=2, label='Medium')
ax5.plot(y_washingSpeed, washingSpeed_fast, 'y', linewidth=2, label='Fast')
ax5.plot(y_washingSpeed, washingSpeed_veryfast, 'm', linewidth=2, label='Very fast')
ax5.set_title('Washing Speed')
ax5.legend()

ax6.plot(y_waterIntake, waterIntake_little, 'r', linewidth=2, label='Little')
ax6.plot(y_waterIntake, waterIntake_normal, 'g', linewidth=2, label='Normal')
ax6.plot(y_waterIntake, waterIntake_alotof, 'b', linewidth=2, label='A lot of')
ax6.set_title('Water Intake')
ax6.legend()

ax7.plot(y_waterTemp, waterTemp_low, 'r', linewidth=2, label='Low')
ax7.plot(y_waterTemp, waterTemp_normal, 'g', linewidth=2, label='Normal')
ax7.plot(y_waterTemp, waterTemp_high, 'b', linewidth=2, label='High')
ax7.set_title('Water Temperature')
ax7.legend()

plt.tight_layout()
```

## Input crisp values

```python
input_dirtinessOfClothes = 50
input_typeOfDirt = 20.8
input_volOfClothes = 28
input_typeOfFabric = 21.5
```

## Fuzzification

```python
dirtinessOfClothes_fit_small = fuzz.interp_membership(x_dirtinessOfClothes, dirtinessOfClothes_small, input_dirtinessOfClothes)
dirtinessOfClothes_fit_med = fuzz.interp_membership(x_dirtinessOfClothes, dirtinessOfClothes_med, input_dirtinessOfClothes)
dirtinessOfClothes_fit_large = fuzz.interp_membership(x_dirtinessOfClothes, dirtinessOfClothes_large, input_dirtinessOfClothes)

typeOfDirt_fit_notgreasy = fuzz.interp_membership(x_typeOfDirt, typeOfDirt_notgreasy, input_typeOfDirt)
typeOfDirt_fit_med = fuzz.interp_membership(x_typeOfDirt, typeOfDirt_med, input_typeOfDirt)
typeOfDirt_fit_greasy = fuzz.interp_membership(x_typeOfDirt, typeOfDirt_greasy, input_typeOfDirt)

volOfClothes_fit_small = fuzz.interp_membership(x_volOfClothes, volOfClothes_small, input_volOfClothes)
volOfClothes_fit_med = fuzz.interp_membership(x_volOfClothes, volOfClothes_med, input_volOfClothes)
volOfClothes_fit_large = fuzz.interp_membership(x_volOfClothes, volOfClothes_large, input_volOfClothes)

typeOfFabric_fit_silk = fuzz.interp_membership(x_typeOfFabric, typeOfFabric_silk, input_typeOfFabric)
typeOfFabric_fit_woolen = fuzz.interp_membership(x_typeOfFabric, typeOfFabric_woolen, input_typeOfFabric)
typeOfFabric_fit_cotton = fuzz.interp_membership(x_typeOfFabric, typeOfFabric_cotton, input_typeOfFabric)
```

## Knowledge base (10 Rules) 

```python
# Rule 1
rule1_antecedent = np.fmin(np.fmin(np.fmin(dirtinessOfClothes_fit_large, typeOfDirt_fit_greasy), volOfClothes_fit_large), typeOfFabric_fit_silk)
rule1 = []             
rule1.append(np.fmin(rule1_antecedent, washingTime_long))
rule1.append(np.fmin(rule1_antecedent, washingSpeed_med))
rule1.append(np.fmin(rule1_antecedent, waterIntake_normal))
rule1.append(np.fmin(rule1_antecedent, waterTemp_normal))

# Rule 2
rule2_antecedent = np.fmin(np.fmin(np.fmin(dirtinessOfClothes_fit_small, typeOfDirt_fit_notgreasy), volOfClothes_fit_small), typeOfFabric_fit_cotton)
rule2 = []             
rule2.append(np.fmin(rule2_antecedent, washingTime_veryshort))
rule2.append(np.fmin(rule2_antecedent, washingSpeed_veryslow))
rule2.append(np.fmin(rule2_antecedent, waterIntake_little))
rule2.append(np.fmin(rule2_antecedent, waterTemp_low))

# Rule 3
rule3_antecedent = np.fmin(np.fmin(np.fmin(dirtinessOfClothes_fit_med, typeOfDirt_fit_med), volOfClothes_fit_med), typeOfFabric_fit_woolen)
rule3 = []             
rule3.append(np.fmin(rule3_antecedent, washingTime_med))
rule3.append(np.fmin(rule3_antecedent, washingSpeed_med))
rule3.append(np.fmin(rule3_antecedent, waterIntake_normal))
rule3.append(np.fmin(rule3_antecedent, waterTemp_normal))

# Rule 4
rule4_antecedent = np.fmin(np.fmin(np.fmin(dirtinessOfClothes_fit_small, typeOfDirt_fit_notgreasy), volOfClothes_fit_small), typeOfFabric_fit_cotton)
rule4 = []             
rule4.append(np.fmin(rule4_antecedent, washingTime_short))
rule4.append(np.fmin(rule4_antecedent, washingSpeed_med))
rule4.append(np.fmin(rule4_antecedent, waterIntake_normal))
rule4.append(np.fmin(rule4_antecedent, waterTemp_low))

# Rule 5
rule5_antecedent = np.fmin(np.fmin(np.fmin(dirtinessOfClothes_fit_large, typeOfDirt_fit_greasy), volOfClothes_fit_large), typeOfFabric_fit_silk)
rule5 = []             
rule5.append(np.fmin(rule5_antecedent, washingTime_verylong))
rule5.append(np.fmin(rule5_antecedent, washingSpeed_veryfast))
rule5.append(np.fmin(rule5_antecedent, waterIntake_alotof))
rule5.append(np.fmin(rule5_antecedent, waterTemp_high))

# Rule 6
rule6_antecedent = np.fmin(np.fmin(np.fmin(dirtinessOfClothes_fit_small, typeOfDirt_fit_notgreasy), volOfClothes_fit_small), typeOfFabric_fit_woolen)
rule6 = []             
rule6.append(np.fmin(rule6_antecedent, washingTime_short))
rule6.append(np.fmin(rule6_antecedent, washingSpeed_slow))
rule6.append(np.fmin(rule6_antecedent, waterIntake_little))
rule6.append(np.fmin(rule6_antecedent, waterTemp_low))

# Rule 7
rule7_antecedent = np.fmin(np.fmin(np.fmin(dirtinessOfClothes_fit_small, typeOfDirt_fit_notgreasy), volOfClothes_fit_small), typeOfFabric_fit_silk)
rule7 = []             
rule7.append(np.fmin(rule7_antecedent, washingTime_short))
rule7.append(np.fmin(rule7_antecedent, washingSpeed_med))
rule7.append(np.fmin(rule7_antecedent, waterIntake_little))
rule7.append(np.fmin(rule7_antecedent, waterTemp_normal))

# Rule 8
rule8_antecedent = np.fmin(np.fmin(np.fmin(dirtinessOfClothes_fit_med, typeOfDirt_fit_notgreasy), volOfClothes_fit_med), typeOfFabric_fit_woolen)
rule8 = []             
rule8.append(np.fmin(rule8_antecedent, washingTime_med))
rule8.append(np.fmin(rule8_antecedent, washingSpeed_fast))
rule8.append(np.fmin(rule8_antecedent, waterIntake_normal))
rule8.append(np.fmin(rule8_antecedent, waterTemp_normal))

# Rule 9
rule9_antecedent = np.fmin(np.fmin(np.fmin(dirtinessOfClothes_fit_med, typeOfDirt_fit_med), volOfClothes_fit_large), typeOfFabric_fit_woolen)
rule9 = []             
rule9.append(np.fmin(rule9_antecedent, washingTime_long))
rule9.append(np.fmin(rule9_antecedent, washingSpeed_fast))
rule9.append(np.fmin(rule9_antecedent, waterIntake_alotof))
rule9.append(np.fmin(rule9_antecedent, waterTemp_high))

# Rule 10
rule10_antecedent = np.fmin(np.fmin(np.fmin(dirtinessOfClothes_fit_med, typeOfDirt_fit_greasy), volOfClothes_fit_large), typeOfFabric_fit_silk)
rule10 = []             
rule10.append(np.fmin(rule10_antecedent, washingTime_verylong))
rule10.append(np.fmin(rule10_antecedent, washingSpeed_fast))
rule10.append(np.fmin(rule10_antecedent, waterIntake_alotof))
rule10.append(np.fmin(rule10_antecedent, waterTemp_high))

rule_list = [rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10]
```

## Aggregation and Defuzzification process 

```python
output_list = [y_washingTime, y_washingSpeed, y_waterIntake, y_waterTemp]
output_name = ["Washing Time", "Washing Speed", "Water Intake", "Water Temperature"]
output_unit = ['min', 'rpm', 'ltr', 'C']

aggregated_list = []
defuzz_list = []
defuz_activation_list = []

for output_idx, output in enumerate(output_list):
  x = rule_list[0]
  for rule_idx in range(1, len(rule_list)):
    if rule_idx==1:
      x = x[output_idx]
    y = rule_list[rule_idx]
    x = np.fmax(x,y[output_idx])
  aggregated_list.append(x)

  real_ans = fuzz.defuzz(output, x, 'centroid')
  ans = round(real_ans,2)
  defuzz_list.append(real_ans)

  defuz_activation_list.append(fuzz.interp_membership(output, x, real_ans))

  print(output_name[output_idx], ": ", ans, output_unit[output_idx])  
```

## Aggregated rules and defuzzified membership functions for each output 

```python
fig, (ax0, ax1, ax2, ax3) = plt.subplots(nrows=4, figsize=(10, 20))

washingTime0 = np.zeros_like(y_washingTime)
ax0.plot(y_washingTime, washingTime_veryshort, 'r', linestyle='--', label='Very Short')
ax0.plot(y_washingTime, washingTime_short, 'g', linestyle='--', label='Short')
ax0.plot(y_washingTime, washingTime_med, 'b', linestyle='--', label='Medium')
ax0.plot(y_washingTime, washingTime_long, 'y', linestyle='--', label='Long')
ax0.plot(y_washingTime, washingTime_verylong, 'm', linestyle='--', label='Very Long')
ax0.fill_between(y_washingTime, washingTime0, aggregated_list[0], facecolor='r', alpha=0.5)
ax0.plot([defuzz_list[0], defuzz_list[0]], [0, defuz_activation_list[0]], 'k', linewidth=1.5, alpha=0.9)
ax0.set_title('Washing Time result')
ax0.legend()

washingSpeed0 = np.zeros_like(y_washingSpeed)
ax1.plot(y_washingSpeed, washingSpeed_veryslow, 'r', linestyle='--', label='Very Slow')
ax1.plot(y_washingSpeed, washingSpeed_slow, 'g', linestyle='--', label='Slow')
ax1.plot(y_washingSpeed, washingSpeed_med, 'b', linestyle='--', label='Medium')
ax1.plot(y_washingSpeed, washingSpeed_fast, 'y', linestyle='--', label='Fast')
ax1.plot(y_washingSpeed, washingSpeed_veryfast, 'm', linestyle='--', label='Very Fast')
ax1.fill_between(y_washingSpeed, washingSpeed0, aggregated_list[1], facecolor='g', alpha=0.5)
ax1.plot([defuzz_list[1], defuzz_list[1]], [0, defuz_activation_list[1]], 'k', linewidth=1.5, alpha=0.9)
ax1.set_title('Washing Speed result')
ax1.legend()

waterIntake0 = np.zeros_like(y_waterIntake)
ax2.plot(y_waterIntake, waterIntake_little, 'r', linestyle='--', label='Little')
ax2.plot(y_waterIntake, waterIntake_normal, 'g', linestyle='--', label='Normal')
ax2.plot(y_waterIntake, waterIntake_alotof, 'b', linestyle='--', label='A lot of')
ax2.fill_between(y_waterIntake, waterIntake0, aggregated_list[2], facecolor='b', alpha=0.5)
ax2.plot([defuzz_list[2], defuzz_list[2]], [0, defuz_activation_list[2]], 'k', linewidth=1.5, alpha=0.9)
ax2.set_title('Water Intake result')
ax2.legend()

waterTemp0 = np.zeros_like(y_waterTemp)
ax3.plot(y_waterTemp, waterTemp_low, 'r', linestyle='--', label='Low')
ax3.plot(y_waterTemp, waterTemp_normal, 'g', linestyle='--', label='Normal')
ax3.plot(y_waterTemp, waterTemp_high, 'b', linestyle='--', label='High')
ax3.fill_between(y_waterTemp, waterTemp0, aggregated_list[3], facecolor='y', alpha=0.5)
ax3.plot([defuzz_list[3], defuzz_list[3]], [0, defuz_activation_list[3]], 'k', linewidth=1.5, alpha=0.9)
ax3.set_title('Water Temperature result')
ax3.legend()

plt.tight_layout()
```
